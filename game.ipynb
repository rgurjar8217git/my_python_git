```python
import random

def game_with_points():
    secret_number = random.randint(1, 10)
    chances = 5
    points = 0  # Starting points

    print(" Number Guessing Game (Points System):")
    print("Rules: Sahi guess pe +10, galat guess pe -2.")
    print("Maine 1 se 10 ke beech ek number socha hai.")

    for i in range(1, chances + 1):
        try:
            guess = int(input(f"\nchance {i} (Points: {points}): Guess karein: "))

            if guess == secret_number:
                points += 10 # Points plus
                print(f"Mubarak ho! Sahi guess. Aapko mile +10 points!")
                print(f"Final Score: {points}")
                return
            else:
                points -= 2 # Points minus
                if guess < secret_number:
                    print("Chhota guess hai! (-2 points)")
                else:
                    print("Bada guess hai! (-2 points)")

        except ValueError:
            print("Invalid input! Sirf numbers use karein.")

    print(f"\nGame Over! Sahi number {secret_number} tha.")
    print(f"Aapka total score: {points}")

if __name__ == "__main__":
    game_with_points()
```
