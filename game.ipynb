```python
import random # Sabse pehle random library ko import karein

# Computer 1 se 100 ke beech ek secret number choose karega
secret_number = random.randint(1, 100)

chance = 3
attempt = 0
print("Welcome! Computer ne 1 se 100 ke beech ek number socha hai.")
print("Aapke paas 3 chances hain:")

while attempt < chance:
    guess = int(input(f"Attempt {attempt + 1}: Guess the number: "))

    if guess == secret_number:
        print(f"Bahut badhiya! Aapne sahi guess kiya. Number {secret_number} hi tha!")
        break
    elif guess < secret_number:
        print("Thoda bada number try karein! (Too Low)")
        attempt = attempt + 1
    else:
        print("Thoda chhota number try karein! (Too High)")
        attempt = attempt + 1

if attempt == chance:
    print(f"You lose! Sahi number {secret_number} tha.")
```
